```python
import pandas as pd
from sklearn.datasets import load_iris
```

```python
stuff = load_iris()
```

```python
dir(stuff)
```

```python
dataset = pd.DataFrame(data=stuff.data,columns=stuff.feature_names)
```

```python
dataset
```

```python
dataset.isnull()
```

```python
dataset["target"] = stuff.target
```

```python
dataset
```

```python
dataset["flower_name"] = dataset.target.apply(lambda x: stuff.target_names[x]) 
```

```python
dataset
```

```python
from sklearn.model_selection import train_test_split
```

```python
x  =dataset.drop(["target","flower_name"],axis="columns")
```

```python
y = dataset["target"]
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y,train_size=0.9)
```

```python
from sklearn.svm import SVC
```

```python
model = SVC(kernel="linear")
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```
